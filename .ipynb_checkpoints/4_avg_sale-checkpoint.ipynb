```python
data = pd.read_csv("data/category/3_avgPrice.csv")
keywords = data["relative_keyword"]
avg_sales_lst = []
```

```python
for idx,keyword in enumerate(keywords):
    
    params = {
    'frm': 'NVSCPRO',
    'pagingIndex': '1',
    'pagingSize': '80',
    'productSet': 'total',
    'query': f"{keyword}",
    'sort': 'rel',
    'viewType': 'list',
    }

    response = requests.get('https://search.shopping.naver.com/api/search/all', params=params, cookies=cookies, headers=headers)
    itemlist = json.loads(response.text)

    sales_lst = []
    cnt = 0
    for idx,product in enumerate(itemlist['shoppingResult']['products']):
        purchaseCnt = product['purchaseCnt'],
        price = product['price'],

        purchaseCnt = purchaseCnt[0]
        price =  int(price[0])

        sales = purchaseCnt * price

    
        if sales > 0 : 
                sales_lst.append(sales)
        
    try:
        avg_sales = int(np.mean(sales_lst)/6)
    except :
        avg_sales = None
    
    avg_sales_lst.append(avg_sales)
    print(f"{keyword}: {avg_sales} ")

```

```python
data["avg_sale"] = pd.Series(avg_sales_lst)
data["avg_sale"]
```
