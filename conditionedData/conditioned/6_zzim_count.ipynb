```python
 """
    찜 ( 10-50-100-1000 ) 
    *광고는 개수에 포함하지 않음*
"""

df_zzim = pd.read_csv("conditionedData/data/6_zzim_count/6_zzim_count_only_requests.csv")

num = 734

for keyword in keywords[num:] :
    
    cnt = 0
    total_zzim = 0 
    
    zzim_10 = 0
    zzim_50 = 0
    zzim_100 = 0
    zzim_1000 = 0
    
    product_params = {
            'adQuery': f'{keyword}',
            'eq': '',
            'frm': 'NVSHATC',
            'iq': '',
            'origQuery': f'{keyword}',
            'pagingIndex': '1',
            'pagingSize': '40',
            'productSet': 'total',
            'query': f'{keyword}',
            'sort': 'rel',
            'viewType': 'list',
            'xq': '',
        }

    response = requests.get('https://search.shopping.naver.com/api/search/all', params=product_params, cookies=product_cookies, headers=product_headers)
    itemlists = json.loads(response.text)

    for idx,product in enumerate(itemlists["shoppingResult"]["products"]):

        id = product["id"]

        # Zzim 개수 카운트
        json_data = {
            'params': [
            {
                'nvMid': f'{id}',
                'chpid': None,
                'isBook': False,
                },
            ]
        }

        response = requests.post(
            'https://search.shopping.naver.com/api/product-zzim/products',
            cookies=zzim_cookies,
            headers=zzim_headers,
            json=json_data,
        )
        zzim_json = json.loads(response.text)
        
        try :
            zzim = zzim_json["zzim"][f"{id}"]["count"]

            # review 개수 나누기  + 평균
            if (zzim > 1000):
                zzim_1000 += 1
            elif (zzim > 100) :  
                zzim_100 += 1
            elif (zzim > 50) :  
                zzim_50 += 1
            else : zzim_10 += 1
        
            total_zzim += zzim
            cnt += 1
        except:
            pass 
        
    avg_zzim = int(total_zzim/cnt)
    
    print(f"{num}.{keyword} : {avg_zzim}")
    print(f"50개이하:{zzim_10} | 50개 이상:{zzim_50} | 100개 이상 : {zzim_100} | 1000개 이상:{zzim_1000}")
    print("----------------------------------------------------------------------")
    
    tmp = pd.DataFrame({
            "relative_keyword":keyword, 
            "zzim_10":[zzim_10],"zzim_50":[zzim_50],"zzim_100":[zzim_100],"zzim_1000":[zzim_1000],
            })    
        
    df_zzim = pd.concat([df_zzim,tmp],axis=0)
    
    num += 1  # keyword indexing 
    
    df_zzim.to_csv("conditionedData/data/6_zzim_count/6_zzim_count_only_requests.csv",index=False)

```
